# 8  **Black_Friday_Sales_Occupation and Product Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("BlackFriday.csv")
```

```python
occupation_purchase = df.groupby('Occupation')['Purchase'].mean().sort_values(ascending=False)
print("\nðŸŽ¯ Average Purchase by Occupation:")
print(occupation_purchase)
```

```python
total_purchase = df.groupby('Occupation')['Purchase'].sum().sort_values(ascending=False)
print("\nðŸ’° Total Purchase by Occupation:")
print(total_purchase)
```

```python
top_product_by_occupation = df.groupby('Occupation')['Product_Category_1'].agg(lambda x: x.mode()[0])
print("\nðŸ“¦ Most Common Product_Category_1 by Occupation:")
print(top_product_by_occupation)
```

```python
occupation_counts = df['Occupation'].value_counts().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=occupation_purchase.index, y=occupation_purchase.values, palette='viridis')
plt.title("Average Purchase by Occupation")
plt.xlabel("Occupation Code")
plt.ylabel("Average Purchase Amount")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Product_Category_1', hue='Occupation', data=df, palette='tab20', edgecolor='gray')
plt.title("Product Category 1 vs Occupation")
plt.xlabel("Product Category 1")
plt.ylabel("Count")
plt.legend(title="Occupation", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
