```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean dataset
df = pd.read_csv("BlackFriday.csv")
df.columns = df.columns.str.strip()
```

```python
# ----- Descriptive Statistics -----
print("ðŸ“ˆ Descriptive Statistics for Numerical Columns:\n")
print(df.describe())
```

```python
print("\nðŸ§¾ Descriptive Statistics for Categorical Columns:\n")
print(df.describe(include='object'))
```

```python
print("\nMean Purchase:", df['Purchase'].mean())
print("Median Purchase:", df['Purchase'].median())
print("Range of Purchase:", df['Purchase'].max() - df['Purchase'].min())
```

```python
print("\nUnique Users:", df['User_ID'].nunique())
print("Unique Products:", df['Product_ID'].nunique())
print("\nMost Frequent Age Group:", df['Age'].mode()[0])
print("Most Purchased Product Category 1:", df['Product_Category_1'].mode()[0])
```

# ----- Visualizations -----


```python
# Set a seaborn style
sns.set(style="whitegrid")

# 1. Distribution of Purchase Amounts
plt.figure(figsize=(8, 5))
sns.histplot(df['Purchase'], bins=50, kde=True)
plt.title("Distribution of Purchase Amounts")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# 2. Average Purchase by Age Group
plt.figure(figsize=(7, 5))
sns.barplot(x='Age', y='Purchase', data=df, estimator='mean', ci=None, palette='Set2')
plt.title("Average Purchase by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Purchase")
plt.tight_layout()
plt.show()
```

```python
# 3. Average Purchase by Gender
plt.figure(figsize=(6, 5))
sns.barplot(x='Gender', y='Purchase', data=df, estimator='mean', ci=None, palette='Set1')
plt.title("Average Purchase by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Purchase")
plt.tight_layout()
plt.show()
```

```python
# 4. Purchase Distribution by Product_Category_1
plt.figure(figsize=(10, 5))
sns.boxplot(x='Product_Category_1', y='Purchase', data=df)
plt.title("Purchase Distribution by Product Category 1")
plt.xlabel("Product Category 1")
plt.ylabel("Purchase Amount")
plt.tight_layout()
plt.show()
```

```python
# 5. Count of Customers by Age Group
plt.figure(figsize=(7, 5))
sns.countplot(x='Age', data=df, palette='coolwarm')
plt.title("Number of Customers by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
