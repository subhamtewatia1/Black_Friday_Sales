# 7  **Black_Friday_Sales_Column Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("BlackFriday.csv")
```

```python
print("\nüßæ Dataset Info:")
print(df.info())
```

```python
print("\n‚ö†Ô∏è Missing Values (count and percent):")
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_table = pd.DataFrame({'Missing Count': missing, 'Percent': missing_percent})
print(missing_table[missing_table['Missing Count'] > 0])
```

```python
print("\nüìä Descriptive Statistics (numerical columns):")
print(df.describe())
```

```python
print("\nüî¢ Unique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
print("\nüìÅ Data Types:")
print(df.dtypes)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("üìå Correlation Heatmap (Numerical Features)")
plt.tight_layout()
plt.show()
```
