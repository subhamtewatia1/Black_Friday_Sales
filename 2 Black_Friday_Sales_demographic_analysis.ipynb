```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean the dataset
df = pd.read_csv("BlackFriday.csv")
df.columns = df.columns.str.strip()
```

```python
# ----- Demographic-Based Purchase Analysis -----

# 1. Average Purchase by Gender
print("üßë‚Äçü§ù‚Äçüßë Average Purchase by Gender:")
print(df.groupby('Gender')['Purchase'].mean(), "\n")

# 2. Average Purchase by Age Group
print("üéÇ Average Purchase by Age Group:")
print(df.groupby('Age')['Purchase'].mean(), "\n")

# 3. Average Purchase by City Category
print("üèôÔ∏è Average Purchase by City Category:")
print(df.groupby('City_Category')['Purchase'].mean(), "\n")

# 4. Average Purchase by Marital Status
print("üíç Average Purchase by Marital Status:")
print(df.groupby('Marital_Status')['Purchase'].mean(), "\n")
```

```python
sns.set(style="whitegrid")
y_range = (5000, 15000)
age_order = ['0-17', '18-25', '26-35', '36-45', '46-50', '51-55', '55+']

plots = [
    ("Gender", "pastel", None, "Average Purchase by Gender"),
    ("Age", "Set2", age_order, "Average Purchase by Age Group"),
    ("City_Category", "coolwarm", None, "Average Purchase by City Category"),
    ("Marital_Status", "muted", None, "Average Purchase by Marital Status")
]

for x, palette, order, title in plots:
    plt.figure(figsize=(6, 5))
    sns.barplot(x=x, y='Purchase', data=df, estimator='mean', ci=None, palette=palette, order=order)
    plt.title(title)
    plt.xlabel(x.replace('_', ' '))
    plt.ylabel("Average Purchase")
    plt.ylim(y_range)
    plt.tight_layout()
    plt.show()
```
